```python
# Import all required libraries
import tensorflow as tf
from tensorflow.keras import layers, models, applications
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
import matplotlib.pyplot as plt
import numpy as np
import os
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
import seaborn as sns
from statsmodels.stats.proportion import proportions_ztest
from collections import Counter

# Set random seeds for reproducibility
tf.random.set_seed(42)
np.random.seed(42)

# Suppress warnings
import logging
logging.getLogger('tensorflow').setLevel(logging.ERROR)

# Define base dataset path
dataset_base = '/kaggle/input/brain-tumor-mri-dataset'

# Check what's in the dataset directory
print("Contents of dataset directory:")
try:
    contents = os.listdir(dataset_base)
    print(contents)
    
    # Find training and testing directories (case-insensitive)
    train_dir = None
    test_dir = None
    
    for item in contents:
        if item.lower() in ['training', 'train']:
            train_dir = os.path.join(dataset_base, item)
        elif item.lower() in ['testing', 'test']:
            test_dir = os.path.join(dataset_base, item)
    
    # If not found, try common variations
    if train_dir is None:
        for item in contents:
            if any(term in item.lower() for term in ['train', 'training']):
                train_dir = os.path.join(dataset_base, item)
                break
                
    if test_dir is None:
        for item in contents:
            if any(term in item.lower() for term in ['test', 'testing']):
                test_dir = os.path.join(dataset_base, item)
                break
    
    # Final check
    if train_dir and os.path.exists(train_dir):
        print(f"Using training directory: {train_dir}")
        print(f"Training subdirectories: {os.listdir(train_dir)}")
    else:
        print("Training directory not found. Available directories:")
        for item in contents:
            item_path = os.path.join(dataset_base, item)
            if os.path.isdir(item_path):
                print(f"{item}: {os.listdir(item_path)[:5]}")  # Show first 5 items
        
        # Manual override if needed
        train_dir = os.path.join(dataset_base, contents[0])  # Use first directory as fallback
    
    if test_dir and os.path.exists(test_dir):
        print(f"Using testing directory: {test_dir}")
        print(f"Testing subdirectories: {os.listdir(test_dir)}")
    else:
        test_dir = train_dir  # Fallback to using training directory for testing
        print(f"Testing directory not found, using training directory: {train_dir}")

except FileNotFoundError:
    print(f"Dataset directory {dataset_base} not found.")
    print("Please check if you've added the correct dataset to your Kaggle notebook.")
    print("Click on '+ Add Data' and search for 'brain-tumor-mri-dataset'")
    # Set placeholder paths
    train_dir = '/kaggle/input/brain-tumor-mri-dataset/Training'
    test_dir = '/kaggle/input/brain-tumor-mri-dataset/Testing'

# Image parameters
IMG_SIZE = (224, 224)
BATCH_SIZE = 32

# Advanced data augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.3,
    height_shift_range=0.3,
    shear_range=0.2,
    zoom_range=0.3,
    horizontal_flip=True,
    vertical_flip=True,
    brightness_range=[0.8, 1.2],
    fill_mode='nearest',
    validation_split=0.15  # Using 15% of training data for validation
)

test_datagen = ImageDataGenerator(rescale=1./255)

# Create data generators with error handling
try:
    # Training data generator
    train_generator = train_datagen.flow_from_directory(
        train_dir,
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        subset='training',
        shuffle=True
    )

    # Validation data generator
    validation_generator = train_datagen.flow_from_directory(
        train_dir,
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        subset='validation',
        shuffle=True
    )

    # Test data generator
    test_generator = test_datagen.flow_from_directory(
        test_dir,
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        shuffle=False
    )

    # Calculate class weights for handling imbalance
    class_weights = compute_class_weight(
        class_weight='balanced',
        classes=np.unique(train_generator.classes),
        y=train_generator.classes
    )
    class_weights = dict(enumerate(class_weights))

    print("Class weights:", class_weights)
    print("Classes:", train_generator.class_indices)
    
except FileNotFoundError as e:
    print(f"Error: {e}")
    print("Please check the dataset paths and try again.")
    # Create dummy generators to prevent further errors
    print("Creating dummy data generators for demonstration...")
    
    # Create a small dummy dataset
    dummy_data = np.random.random((32, 224, 224, 3))
    dummy_labels = tf.keras.utils.to_categorical(np.random.randint(0, 4, 32), 4)
    
    # Create dummy generators
    class DummyGenerator:
        def __init__(self):
            self.batch_size = BATCH_SIZE
            self.num_classes = 4
            self.classes = np.random.randint(0, 4, 100)
            self.class_indices = {'glioma': 0, 'meningioma': 1, 'notumor': 2, 'pituitary': 3}
            
        def __iter__(self):
            return self
        
        def __next__(self):
            x = np.random.random((self.batch_size, 224, 224, 3))
            y = tf.keras.utils.to_categorical(np.random.randint(0, self.num_classes, self.batch_size), self.num_classes)
            return x, y
            
        def reset(self):
            pass
    
    train_generator = DummyGenerator()
    validation_generator = DummyGenerator()
    test_generator = DummyGenerator()
    class_weights = {0: 1, 1: 1, 2: 1, 3: 1}

# Build VGG16 Model with Robust Weight Loading
# First, try to enable internet access programmatically (may not work in Kaggle)
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'  # Reduce TensorFlow logging

# Try multiple approaches to load VGG16 weights
try:
    # Approach 1: Try to download from internet (requires internet enabled)
    print("Attempting to load VGG16 with ImageNet weights...")
    base_model = applications.VGG16(
        weights='imagenet',
        include_top=False,
        input_shape=(224, 224, 3)
    )
    print("Successfully loaded VGG16 with ImageNet weights")
    
except Exception as e:
    print(f"Failed to download weights: {e}")
    
    # Approach 2: Try to use Kaggle's cached weights if available
    try:
        # Check if weights are cached in Kaggle's environment
        cache_path = '/root/.keras/models/vgg16_weights_tf_dim_ordering_tf_kernels_notop.h5'
        if os.path.exists(cache_path):
            print("Loading VGG16 weights from Kaggle cache...")
            base_model = applications.VGG16(
                weights=None,
                include_top=False,
                input_shape=(224, 224, 3)
            )
            base_model.load_weights(cache_path)
            print("Successfully loaded VGG16 weights from Kaggle cache")
        else:
            raise FileNotFoundError("Cached weights not found")
            
    except Exception as e2:
        print(f"Failed to load cached weights: {e2}")
        
        # Approach 3: Try to use pre-downloaded weights from Kaggle datasets
        try:
            # Common paths for VGG16 weights in Kaggle datasets
            possible_paths = [
                '/kaggle/input/vgg16/vgg16_weights_tf_dim_ordering_tf_kernels_notop.h5',
                '/kaggle/input/keras-pretrained-models/vgg16_weights_tf_dim_ordering_tf_kernels_notop.h5',
                '/kaggle/input/vgg16-weights/vgg16_weights_tf_dim_ordering_tf_kernels_notop.h5'
            ]
            
            weights_path = None
            for path in possible_paths:
                if os.path.exists(path):
                    weights_path = path
                    break
            
            if weights_path:
                print(f"Loading VGG16 weights from dataset: {weights_path}")
                base_model = applications.VGG16(
                    weights=None,
                    include_top=False,
                    input_shape=(224, 224, 3)
                )
                base_model.load_weights(weights_path)
                print("Successfully loaded VGG16 weights from dataset")
            else:
                raise FileNotFoundError("No weight files found in datasets")
                
        except Exception as e3:
            print(f"Failed to load weights from dataset: {e3}")
            
            # Approach 4: Last resort - use random weights
            print("Initializing VGG16 with random weights (performance will be lower)")
            base_model = applications.VGG16(
                weights=None,
                include_top=False,
                input_shape=(224, 224, 3)
            )

# Freeze the base model
base_model.trainable = False
print("Base model frozen")

# Create custom top layers
model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(512, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.5),
    layers.Dense(256, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.3),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(train_generator.num_classes, activation='softmax')
])

# Compile the model
model.compile(
    optimizer=Adam(learning_rate=0.0001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Display model summary
print("Model architecture:")
model.summary()

# If using random weights, note that more training may be needed
if hasattr(base_model, 'weights') and len(base_model.weights) > 0:
    if base_model.weights[0].numpy().sum() == 0:
        print("\n⚠️  WARNING: Using randomly initialized weights.")
        print("   Model performance will be lower than with pre-trained weights.")
        print("   You may need to train for more epochs or use a different approach.")
else:
    print("\n⚠️  WARNING: Using randomly initialized weights.")
    print("   Model performance will be lower than with pre-trained weights.")

# Callbacks
early_stop = EarlyStopping(monitor='val_accuracy', patience=10, restore_best_weights=True, mode='max')
reduce_lr = ReduceLROnPlateau(monitor='val_accuracy', factor=0.5, patience=5, min_lr=1e-7, mode='max')
checkpoint = ModelCheckpoint('best_model_phase1.h5', monitor='val_accuracy', 
                            save_best_only=True, mode='max', verbose=1)

# Train the model (Phase 1 - Feature Extraction)
print("Starting Phase 1 Training (Feature Extraction)")
history_phase1 = model.fit(
    train_generator,
    epochs=30,
    validation_data=validation_generator,
    callbacks=[early_stop, reduce_lr, checkpoint],
    class_weight=class_weights
)

# Unfreeze some layers for fine-tuning
base_model.trainable = True
# Freeze the first 10 layers and unfreeze the rest
for layer in base_model.layers[:10]:
    layer.trainable = False

# Recompile with a lower learning rate
model.compile(
    optimizer=Adam(learning_rate=0.00001),
    loss='categorical_crossentropy',
    metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall()]
)

# New callbacks for fine-tuning
checkpoint_ft = ModelCheckpoint('best_model_phase2.h5', monitor='val_accuracy', 
                               save_best_only=True, mode='max', verbose=1)

print("Starting Phase 2 Training (Fine-Tuning)")
history_phase2 = model.fit(
    train_generator,
    epochs=20,
    validation_data=validation_generator,
    callbacks=[early_stop, reduce_lr, checkpoint_ft],
    class_weight=class_weights
)

# Load the best model
model.load_weights('best_model_phase2.h5')

# Evaluate on test data
test_loss, test_accuracy, test_precision, test_recall = model.evaluate(test_generator)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
print(f"Test Precision: {test_precision * 100:.2f}%")
print(f"Test Recall: {test_recall * 100:.2f}%")

# Generate predictions
test_generator.reset()
predictions = model.predict(test_generator, verbose=1)
predicted_classes = np.argmax(predictions, axis=1)

# True classes
true_classes = test_generator.classes
class_labels = list(test_generator.class_indices.keys())

# Classification report
print("Classification Report:")
print(classification_report(true_classes, predicted_classes, target_names=class_labels))

# Confusion matrix
cm = confusion_matrix(true_classes, predicted_classes)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_labels, yticklabels=class_labels)
plt.title('Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()

# Combine training histories
def combine_histories(history1, history2):
    combined = {}
    for key in history1.history.keys():
        combined[key] = history1.history[key] + history2.history[key]
    return combined

combined_history = combine_histories(history_phase1, history_phase2)

# Plot training history
def plot_history(history):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
    
    # Plot accuracy
    ax1.plot(history['accuracy'], label='Training Accuracy')
    ax1.plot(history['val_accuracy'], label='Validation Accuracy')
    ax1.set_title('Model Accuracy')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Accuracy')
    ax1.legend()
    
    # Plot loss
    ax2.plot(history['loss'], label='Training Loss')
    ax2.plot(history['val_loss'], label='Validation Loss')
    ax2.set_title('Model Loss')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Loss')
    ax2.legend()
    
    plt.tight_layout()
    plt.show()

plot_history(combined_history)

# Calculate F1-score
from sklearn.metrics import precision_recall_fscore_support
precision, recall, f1, _ = precision_recall_fscore_support(true_classes, predicted_classes, average='weighted')
print(f"Weighted F1-Score: {f1:.4f}")

# Calculate and print additional metrics
accuracy = np.mean(true_classes == predicted_classes)
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")

# Statistical significance testing
n_samples = len(true_classes)
null_accuracy = 0.25  # Random chance accuracy for 4 classes

# Perform one-sample proportion test
count_success = int(test_accuracy * n_samples)
z_stat, p_value = proportions_ztest(
    count=count_success,
    nobs=n_samples,
    value=null_accuracy,
    alternative='larger'  # One-tailed test (we expect better than random)
)

# Function to determine significance level
def get_significance_level(p_val):
    if p_val < 0.05:
        return "statistically significant"
    elif 0.05 <= p_val < 0.1:
        return "marginally significant"
    else:
        return "not statistically significant"

print(f"P-value compared to random chance (25%): {p_value:.4e}")
print(f"The model's performance is {get_significance_level(p_value)}")

# Compare against majority class baseline
class_distribution = Counter(true_classes)
majority_class_count = max(class_distribution.values())
majority_baseline_accuracy = majority_class_count / n_samples

count_success_majority = int(test_accuracy * n_samples)
z_stat_majority, p_value_majority = proportions_ztest(
    count=count_success_majority,
    nobs=n_samples,
    value=majority_baseline_accuracy,
    alternative='larger'
)

print(f"\nMajority class baseline accuracy: {majority_baseline_accuracy:.3f}")
print(f"P-value compared to majority baseline: {p_value_majority:.4e}")
print(f"Significance against majority baseline: {get_significance_level(p_value_majority)}")

# Hyperparameter tracking and results compilation
hyperparameters = {
    'Learning Rate Phase 1': 0.0001,
    'Learning Rate Phase 2': 0.00001,
    'Batch Size': BATCH_SIZE,
    'Hidden Sizes': [512, 256, 128],
    'Epochs Phase 1': len(history_phase1.history['accuracy']),
    'Epochs Phase 2': len(history_phase2.history['accuracy']),
    'Total Epochs': len(history_phase1.history['accuracy']) + len(history_phase2.history['accuracy'])
}

# Calculate additional metrics
from sklearn.metrics import cohen_kappa_score, f1_score, precision_score, recall_score

kappa_score = cohen_kappa_score(true_classes, predicted_classes)
weighted_f1 = f1_score(true_classes, predicted_classes, average='weighted')
weighted_precision = precision_score(true_classes, predicted_classes, average='weighted')
weighted_recall = recall_score(true_classes, predicted_classes, average='weighted')

# Create results table
results_table = """
=== HYPERPARAMETERS ===
Learning Rate (Phase 1): {lr1}
Learning Rate (Phase 2): {lr2}
Batch Size: {batch_size}
Hidden Sizes: {hidden_sizes}
Epochs (Phase 1): {epochs1}
Epochs (Phase 2): {epochs2}
Total Epochs: {total_epochs}

=== RESULTS ===
Accuracy: {accuracy:.2f}%
Precision: {precision:.2f}%
Recall: {recall:.2f}%
F1 Score: {f1:.2f}%
Cohen's Kappa Score: {kappa:.4f}
P-value (vs random): {p_value:.4e}
Significance (vs random): {significance_random}
P-value (vs majority): {p_value_majority:.4e}
Significance (vs majority): {significance_majority}

=== CLASS-WISE METRICS ===
{class_report}
""".format(
    lr1=hyperparameters['Learning Rate Phase 1'],
    lr2=hyperparameters['Learning Rate Phase 2'],
    batch_size=hyperparameters['Batch Size'],
    hidden_sizes=hyperparameters['Hidden Sizes'],
    epochs1=hyperparameters['Epochs Phase 1'],
    epochs2=hyperparameters['Epochs Phase 2'],
    total_epochs=hyperparameters['Total Epochs'],
    accuracy=test_accuracy * 100,
    precision=weighted_precision * 100,
    recall=weighted_recall * 100,
    f1=weighted_f1 * 100,
    kappa=kappa_score,
    p_value=p_value,
    significance_random=get_significance_level(p_value),
    p_value_majority=p_value_majority,
    significance_majority=get_significance_level(p_value_majority),
    class_report=classification_report(true_classes, predicted_classes, target_names=class_labels)
)

print(results_table)

# Save the final model
model.save('brain_tumor_vgg16_final.h5')

# Save predictions and metrics
import pandas as pd
results_df = pd.DataFrame({
    'True_Class': true_classes,
    'Predicted_Class': predicted_classes
})
results_df.to_csv('prediction_results.csv', index=False)

# Print final results
print("=== FINAL RESULTS ===")
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
print(f"Precision: {test_precision * 100:.2f}%")
print(f"Recall: {test_recall * 100:.2f}%")
print(f"F1-Score: {f1 * 100:.2f}%")
print(f"P-value (vs random): {p_value:.4e}")
print(f"Significance: {get_significance_level(p_value)}")
print(f"P-value (vs majority): {p_value_majority:.4e}")
print(f"Significance: {get_significance_level(p_value_majority)}")
```
